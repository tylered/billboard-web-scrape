```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
req = requests.get('https://www.billboard.com/charts/hot-100/')
```

```python
soup = BeautifulSoup(req.text, 'html.parser')
```

```python
# for chart data
song_data = soup.find('div', class_='chart-results-list').find_all('div', class_='o-chart-results-list-row-container')
```

```python
# for header
header_data = soup.find('div', class_='chart-results-list').find('div', 'chart-results-list-header')
```

```python
# get header names
headers = []
for i in header_data.find_all('span'):
    if len(i.text) != 1:
        headers.append(i.text.strip())
```

```python
# remove award and add artist and song title
headers[1] = 'Artist'
headers.insert(2, 'Song Title')
```

```python
# create list to store song data
records = []

# loop through rows to get data
for i in song_data:
    
    # get song titles
    song = i.find_all('h3')
    title = [j.text.strip() for j in song]
    title = title[0]
    
    # get current position, artist, last week position, peak position, weeks on chart
    y = i.find_all('span')
    if y[1].text.strip() == 'NEW' or y[1].text.strip() == 'RE-\nENTRY':
        records.append([y[0].text.strip(), y[3].text.strip(), title, y[7].text.strip(), y[8].text.strip(), y[9].text.strip()])
    else:
        records.append([y[0].text.strip(), y[1].text.strip(), title, y[5].text.strip(), y[6].text.strip(), y[7].text.strip()])
```
